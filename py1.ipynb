```python
import numpy as np
import pandas as pd
from sklearn.impute import SimpleImputer
```

```python
data = {
    "Age" : [25,30,np.nan,45,40,np.nan],
    "Salary" : [50000, 60000, 55000, np.nan, 65000, 62000],
    "Department" : ["HR", "IT", "IT", np.nan, "Finance", "IT"]
}
```

```python
df = pd.DataFrame(data)
```

```python
print(df)
```

```python
df.head()
```

```python
df.isna().sum()
```

```python
df_fill = df.copy()
## numeric columns --> mean imputation
df_fill["Age"] = df_fill["Age"].fillna(df_fill["Age"].mean())
df_fill["Salary"] = df_fill["Salary"].fillna(df_fill["Salary"].mean())
```

```python
# Categorical column → mode 
df_fill["Department"] = df_fill["Department"].fillna(df_fill["Department"].mode()[0])
```

```python
print(df_fill)
```

```python
## using imputer -->>
```

```python
df_imputed = df.copy()
```

```python
# Numeric imputation → mean 
num_imputer = SimpleImputer(strategy="mean") 
# Categorical imputation → most_frequent 
cat_imputer = SimpleImputer(strategy="most_frequent")
```

```python
df_imputed[["Age", "Salary"]] = num_imputer.fit_transform( df_imputed[["Age", "Salary"]])

df_imputed[["Department"]] = cat_imputer.fit_transform(df_imputed[["Department"]])
```

```python
print(df_imputed)
```
